```python
import numpy as np
import pandas as pd
import math 
import matplotlib.pyplot as plt

from sklearn import datasets,mixture
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
df=[]
df = pd.read_csv('data.tsv',sep="\t")
print(df)

plt.ylabel('Waiting time')
plt.xlabel('Eruption time')
plt.plot(df['eruptions'],df['waiting'],linestyle='',marker='o')
print(df)

#Resources: https://scikit-learn.org/stable/modules/generated/sklearn.mixture.GaussianMixture.html#sklearn.mixture.GaussianMixture.fit
#https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html
```

```python
#initialize
iter=10
iter2=70
dfn=pd.DataFrame()
dfn2=pd.DataFrame()
colors = ["navy","darkorange"]
plt.rcParams["figure.figsize"] = (15,18)


#Build model for iter1
GM = GaussianMixture(n_components=2, init_params='random',max_iter=iter)
KM = KMeans(n_clusters=2,init='random',n_init=1,max_iter=iter)

kmfit = KM.fit(df)
kmcenters=kmfit.cluster_centers_
dfn['kmlabels']=kmfit.labels_
print(iter)
print('K-Means')
#print(kmfit.labels_)
print(kmcenters)


gmfit = GM.fit(df)
gmcenters = gmfit.means_
dfn['gmlabels']=gmfit.predict(df)
print(df)
print('Gaussian Mixture')
print(gmfit.weights_)
print(gmcenters)
print('')
print('')

plt.subplot(2,2,1)
for i in range(0,len(df)):
    plt.plot(df['eruptions'][i],df['waiting'][i],linestyle='',
             marker='o',color=colors[dfn['kmlabels'][i]])
plt.plot(kmcenters[0,0],kmcenters[0,1],linestyle='',marker='x',color='red',markersize=16)
plt.plot(kmcenters[1,0],kmcenters[1,1],linestyle='',marker='x',color='red',markersize=16)
plt.ylabel('Waiting time')
plt.xlabel('Eruption time, K-Means clustering, %s iter' % iter)


plt.subplot(2,2,3)
for i in range(0,len(df)):
    plt.plot(df['eruptions'][i],df['waiting'][i],linestyle='',
             marker='o',color=colors[dfn['gmlabels'][i]])
plt.plot(gmcenters[0,0],gmcenters[0,1],linestyle='',marker='x',color='red',markersize=16)
plt.plot(gmcenters[1,0],gmcenters[1,1],linestyle='',marker='x',color='red',markersize=16)
plt.ylabel('Waiting time')
plt.xlabel('Eruption time, GM/EM clustering, %s iter' % iter)
```

```python

#Build model for iter2
GM2 = GaussianMixture(n_components=2, init_params='random',max_iter=iter2)
KM2 = KMeans(n_clusters=2,max_iter=iter2)

kmfit2 = KM2.fit(df)
kmcenters2=kmfit2.cluster_centers_
dfn2['kmlabels']=kmfit2.labels_

print(iter2)
print('K-Means')
#print(kmfit.labels_)
print(kmcenters2)

print('')

gmfit2 = GM2.fit(df)
gmcenters2 = gmfit2.means_
dfn2['gmlabels']=gmfit2.predict(df)

print('Gaussian Mixture')
print(gmfit2.weights_)
print(gmcenters2)
print('')
print('')


plt.subplot(2,2,2)
for i in range(0,len(df)):
    plt.plot(df['eruptions'][i],df['waiting'][i],linestyle='',
             marker='o',color=colors[dfn2['kmlabels'][i]])
plt.plot(kmcenters2[0,0],kmcenters2[0,1],linestyle='',marker='x',color='red',markersize=16)
plt.plot(kmcenters2[1,0],kmcenters2[1,1],linestyle='',marker='x',color='red',markersize=16)
plt.ylabel('Waiting time')
plt.xlabel('Eruption time, K-Means clustering, %s iter' % iter2)


plt.subplot(2,2,4)
for i in range(0,len(df)):
    plt.plot(df['eruptions'][i],df['waiting'][i],linestyle='',
             marker='o',color=colors[dfn2['gmlabels'][i]])
plt.plot(gmcenters2[0,0],gmcenters2[0,1],linestyle='',marker='x',color='red',markersize=16)
plt.plot(gmcenters2[1,0],gmcenters2[1,1],linestyle='',marker='x',color='red',markersize=16)
plt.ylabel('Waiting time')
plt.xlabel('Eruption time, GM/EM clustering, %s iter' % iter2)
```
